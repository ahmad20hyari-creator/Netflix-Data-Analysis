Q.5 What is the average duration of movies and TV shows?
movies=

*   movies= 99.56499755341706
*   TV shows= 1.764947683109118



```python
df['Duration']=df['Duration'].str.replace(' min','')
df['Duration']=df['Duration'].str.replace(' Seasons','')
df['Duration']=df['Duration'].str.replace(' Season','')
df['Duration']=df['Duration'].astype(str).astype(int)
df['Duration']
```

```python
movies_duration=df[df["Type"]=="Movie"]
tv_show_duration=df[df["Type"]=="TV Show"]
```

```python
print(tv_show_duration['Duration'].mean())
print(movies_duration['Duration'].mean())
```

```python
sns.set(style="darkgrid")
sns.kdeplot(data=movies_duration['Duration'], shade=True)

mean_duration = movies_duration['Duration'].mean()

plt.axvline(mean_duration, color='r', linestyle='--', label=f'Mean: {mean_duration:.2f}')

plt.title('Average duration of movies')
plt.xlabel('Duration(min)')
plt.ylabel('Density')
plt.legend()


plt.show()
```

```python
sns.set(style="darkgrid")
sns.kdeplot(data=tv_show_duration['Duration'], shade=True)

mean_duration = tv_show_duration['Duration'].mean()

plt.axvline(mean_duration, color='r', linestyle='--', label=f'Mean: {mean_duration:.2f}')

plt.title('Average duration of TV_show')
plt.xlabel('Duration(Season)')
plt.ylabel('Density')
plt.legend()


plt.show()
```
